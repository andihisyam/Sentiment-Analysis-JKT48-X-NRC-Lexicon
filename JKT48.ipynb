```python
#Preprocessing
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
slangs={'yg':'yang', 'tdk':'tidak', 'pd':'pada', 'mlh':'malah',
 'jgn':'jangan', 'jg':'juga', 'tp':'tapi', 'blkg': 'belakang',
 'dr':'dari', 'klo':'kalo', 'lg':'lagi','ajg':'anjing'}
processed_comments = []
for sentence in gabungan['full_text']:
 # Remove all the special characters
 processed_comment = re.sub(r'\W', ' ', str(sentence))

 # Converting to Lowercase
 processed_comment = processed_comment.lower()

 #Remove number
 processed_comment = re.sub(r'\d+', ' ', processed_comment)
 # remove all single characters
 processed_comment = re.sub(r'\s+[a-zA-Z]\s+', ' ', processed_comment)

 #remove duplicate character
 pattern=reg.compile(r"(.)\1{1,}",reg.DOTALL)
 processed_comment=pattern.sub(r"\1",processed_comment)
 #Corrected Slang words
 words = processed_comment.split()
 rfrm=[slangs[word] if word in slangs else word for word in words]
 processed_comment= " ".join(rfrm)
 #remove stopword
 factory = StopWordRemoverFactory()
 more_stopword = ['tak', 'jd', 'per','nya'] #menambahkan stopword
 stopwords = factory.get_stop_words() + more_stopword
 temp = [t for t in re.findall(r'\b[a-z]+-?[a-z]+\b',processed_comment) if t not in stopwords]
 processed_comment = ' '.join(temp)

 #stemming
 stemmer = StemmerFactory().create_stemmer()
 processed_comment = stemmer.stem(processed_comment)

 # Substituting multiple spaces with single space
 processed_comment = re.sub(r'\s+', ' ', processed_comment, flags=re.I)
 processed_comments.append(processed_comment)
```

```python
#LOAD DICTIONARY
dictionary=pd.read_excel('NRC.xlsx', sheet_name='dict')
dictionary
```

```python
# Load NRC to memory
positive = []
negative = []

for i in range(1, len(dictionary)):
    kata = dictionary.iloc[i, 1]
    if dictionary.iloc[i, 6] == 1:
        positive.append(kata)
    if dictionary.iloc[i, 7] == 1:
        negative.append(kata)

processed_comments = dataset['CleanReview']
datasetangkapos = []

for kalimat in processed_comments:
    baris = []
    baris.append(kalimat)

    f_sentimenpositif = 0
    f_sentimennegatif = 0

    for teks in kalimat.split():
        for j in range(len(positive)):
            if teks == positive[j]:
                f_sentimenpositif += 1
        for j in range(len(negative)):
            if teks == negative[j]:
                f_sentimennegatif += 1

    # Menambahkan aturan jika terdapat kata "kontol", maka langsung diklasifikasikan sebagai sentimen negatif
    if "kontol" in kalimat.lower():
        label_sentimen = 'Negatif'
    # Menentukan sentimen berdasarkan perbandingan jumlah kata positif dan negatif
    elif f_sentimenpositif > f_sentimennegatif:
        label_sentimen = 'Positif'
    elif f_sentimenpositif < f_sentimennegatif:
        label_sentimen = 'Negatif'
    else:
        label_sentimen = 'Netral'

    baris.append(label_sentimen)
    datasetangkapos.append(baris)

# Membuat DataFrame dari dataset yang telah diberi label
dataset_labelled = pd.DataFrame(datasetangkapos, columns=['Review', 'Sentimen'])
```

## Menghting Proporsi Emosi

```python
def predict_emotion(review):
    # Cek apakah kata 'kontol' ada dalam review
    if 'kontol' in review:
        return 'Anger'

    # Inisialisasi skor emosi
    emotion_scores = {
        'Anger': 0,
        'Anticipation': 0,
        'Disgust': 0,
        'Fear': 0,
        'Joy': 0,
        'Sadness': 0,
        'Surprise': 0,
        'Trust': 0
    }

    # Iterasi melalui setiap kata dalam review
    for word in review.split():
        # Periksa apakah kata tersebut ada dalam kamus NRC
        if word in dictionary['Indonesian (id)'].values:
            # Periksa label emosi untuk kata tersebut dalam bahasa Indonesia
            for emotion in emotion_scores.keys():
                if dictionary.loc[dictionary['Indonesian (id)'] == word, emotion].values[0] == 1:
                    emotion_scores[emotion] += 1
        elif word in dictionary['Javanese (jw)'].values:
            # Periksa label emosi untuk kata tersebut dalam bahasa Jawa
            for emotion in emotion_scores.keys():
                if dictionary.loc[dictionary['Javanese (jw)'] == word, emotion].values[0] == 1:
                    emotion_scores[emotion] += 1

    # Tentukan emosi dominan berdasarkan skor tertinggi
    dominant_emotion = max(emotion_scores, key=emotion_scores.get)

    # Jika tidak ada kata yang ditemukan, default prediksi ke Netral
    if all(value == 0 for value in emotion_scores.values()):
        dominant_emotion = 'None'

    return dominant_emotion
```

```python
# Contoh penggunaan fungsi untuk memprediksi emosi pada setiap entri di review
# Misalnya, Anda memiliki DataFrame 'dataset' dengan kolom 'Review'
dataset['Emosi'] = dataset['CleanReview'].apply(predict_emotion)

# Tampilkan DataFrame 'dataset' dengan kolom 'Emosi' yang sudah diprediksi
print(dataset[['CleanReview', 'Emosi']])

```

```python
import matplotlib.pyplot as plt

# Hitung jumlah kemunculan setiap emosi
emotion_counts = dataset['Emosi'].value_counts()

# Buat bar chart
plt.figure(figsize=(10, 6))
ax = emotion_counts.plot(kind='bar', color='skyblue')
plt.title('Proporsi Emosi dalam Dataset')
plt.xlabel('Emosi')
plt.ylabel('Jumlah')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Tambahkan label nilai di atas setiap batang bar
for i, count in enumerate(emotion_counts):
    ax.text(i, count + 10, str(count), ha='center', va='bottom', fontsize=10)

plt.show()

```

## Menentukan topik yang paling sering dibahas

```python
import spacy
import gensim
from gensim import corpora

# Pra-pemrosesan teks dengan spaCy
nlp = spacy.load('en_core_web_sm')
processed_texts = [nlp(text.lower()) for text in dataset['CleanReview']]

# Menghapus stopwords dan simpan token-token yang merupakan kata benda, kata kerja, dan kata sifat
processed_texts = [[token.lemma_ for token in doc if not token.is_stop and token.pos_ in ['NOUN', 'VERB', 'ADJ']] for doc in processed_texts]

# Membangun kamus (dictionary) dari teks yang telah diproses
dictionary = corpora.Dictionary(processed_texts)

# Membuat representasi vektor dari teks menggunakan Bag-of-Words
corpus = [dictionary.doc2bow(text) for text in processed_texts]

# Melatih model LDA
lda_model = gensim.models.LdaModel(corpus, num_topics=5, id2word=dictionary, passes=15)

# Menampilkan topik-topik utama
for idx, topic in lda_model.print_topics():
    print('Topik', idx+1, ':', topic)
```

- Topik pertama dengan persentase tertinggi (28,8%) mencakup tweet yang berisi ucapan selamat dan semangat kepada anggota JKT48, dengan kata kunci seperti ‘‘selamat‘‘, ‘‘semangat‘‘, dan ‘‘member‘‘ yang sering muncul.
- Topik kedua (25,8%) berfokus pada acara langsung dan kegiatan harian anggota, dengan kata kunci "live", "graduation", dan "morning".
- Topik ketiga(22,1%), menunjukkan antusiasme penggemar dalam mengikuti berbagai acara yang diadakan oleh JKT48, seperti yang terlihat dari kata kunci "ikut", "event", dan "jadwal".
- Topik keempat (12,8%) merayakan momen positif dengan kata kunci "birthday", "pap", dan "good".

- Topik kelima (10,5%) berhubungan dengan waktu dan aktivitas anggota, mencerminkan perhatian penggemar terhadap keseharian mereka melalui kata kunci seperti "hour", "member", dan "Freya".
